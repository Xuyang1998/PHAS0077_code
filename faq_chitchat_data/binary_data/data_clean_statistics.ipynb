# Replace Brands name with random ID

```python
def getRandomSet(bits):
    num_set = [chr(i) for i in range(48,58)]
    char_set = [chr(i) for i in range(97,123)]
    total_set = num_set + char_set
    
    value_set = "".join(random.sample(total_set, bits))
    
    return value_set
```

```python
def text_to_length_statistic(path):
  with open(path ,'r') as file:
    data = []
    filecontent = file.readlines()
    for line in filecontent:
      text = line.replace('- ', '')
      text = text.replace('\n', '')
      data.append(text)

  length_list = []
  for i in range(len(data)):
    text_word = data[i].split(' ')
    length = len(text_word)
    length_list.append(length)

  data_dict = {}
  from itertools import groupby
  for k, g in groupby(sorted(length_list), key=lambda x: x//10):
    #a = len(list(g))
    # print('{}-{}: {}'.format(k*5, (k+1)*5-1, len(list(g))))
    data_dict['{}-{}'.format(k*10, (k+1)*10-1)] = len(list(g))
  return data_dict 

```

# Frequency of words

```python
def count_words(s, n):

 
    s_list = s.lower().split(' ')  # ÂçïËØçÁªü‰∏ÄËΩ¨Êç¢‰∏∫Â∞èÂÜôÂΩ¢ÂºèÔºåÂπ∂‰ª•Á©∫Ê†ºËøõË°åÂàáÂàÜ
 
    # ÁªüËÆ°Â≠óÁ¨¶‰∏≤s‰∏≠ÊØè‰∏™ÂçïËØçÂá∫Áé∞ÁöÑÊ¨°Êï∞
    top_n_dict = {}
    for word in s_list:
        if word in top_n_dict:
            top_n_dict[word] += 1
        else:
            top_n_dict[word] = 1
 
    # ÊåâÁÖßÂá∫Áé∞È¢ëÊ¨°ÂØπÂçïËØçËøõË°åÊéíÂ∫èÔºåÂ¶ÇÊûúÂá∫Áé∞È¢ëÊ¨°Áõ∏ÂêåÔºåÂàôÊåâÂ≠óÊØçÈ°∫Â∫èÊéíÂ∫è
    word_frequency = []
    values = sorted(list(set(top_n_dict.values())), reverse=True)  # ÁªüËÆ°ÊâÄÊúâÂçïËØçÂá∫Áé∞ÁöÑÈ¢ëÊ¨°ÊÉÖÂÜµÔºåÂ∞ÜÈ¢ëÊ¨°ÈôçÂ∫èÊîæÂÖ•ÂàóË°®
    for w in values:
        # Â∞ÜÂá∫Áé∞È¢ëÊ¨°Áõ∏ÂêåÁöÑÂçïËØçÊîæÂú®‰∏Ä‰∏™ÂàóË°®Èáå
        word_list = []
        for k, v in top_n_dict.items():
            if v == w:
                word_list.append((k, v))
        # Â∞ÜÂá∫Áé∞È¢ëÊ¨°Áõ∏ÂêåÁöÑÂçïËØçÊéíÂ∫èÂêéÊ∑ªÂä†Âà∞ËØçÈ¢ëÂàóË°®
        word_frequency.extend(sorted(word_list))
 
    # ËøîÂõûÂá∫Áé∞È¢ëÊ¨°ÊéíÂâçnÁöÑÂçïËØç
    return word_frequency[:n]
 
 
# if __name__ == '__main__':
#     print(count_words("Cat bat mat cat bat Cat", 4))
#     print(count_words("betty bought a bit of Butter but the Butter was bitter", 4))
```
