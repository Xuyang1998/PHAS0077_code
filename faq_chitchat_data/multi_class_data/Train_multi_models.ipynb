# Training **Data**

#1. Train and Test data: #
  a. **FAQ**: 333 original examples, 67 of them in each fold test set, and augment augment another 20 T5 examples for rest 266 examples in training set. Thus we have nearly ~5000 examples for each fold's training set.

  b. **Chitchat**: ~5000 examples in total. 80% twitter data, which exclude URL links and preserve sentences and emoji, 20% for ParlAi data. Then, we also selecet appropriate 333 examples which is same to FAQ, twitter 80%, ParlAis 20% for cross-validation. Each fold we choose 67 samples in the 333 selected set for testing and put rest of them in the training set.

## ***Training Results***

```python
import json

accuracy_DIET = []
accuracy_DIET_Pretrained = []
accuracy_BERT = []

precision_DIET = []
precision_DIET_Pretrained = []
precision_BERT = []

recall_DIET = []
recall_DIET_Pretrained = []
recall_BERT = []

f1_DIET = []
f1_DIET_Pretrained = []
f1_BERT = []

# DIET
for i in range(1,6):
  with open('faq_chitchat_data/binary_data/binary_models_results/cv{}_DIET_result/intent_report.json'.format(i)) as f:
    data = json.load(f)
    accuracy_DIET.append(data['accuracy'])
    precision_DIET.append(data['macro avg']['precision'])
    recall_DIET.append(data['macro avg']['recall'])
    f1_DIET.append(data['macro avg']['f1-score'])

# DIET Pretrained
for i in range(1,6):
  with open('faq_chitchat_data/binary_data/binary_models_results/cv{}_DIET_PreTrained_result/intent_report.json'.format(i)) as f:
    data = json.load(f)
    accuracy_DIET_Pretrained.append(data['accuracy'])
    precision_DIET_Pretrained.append(data['macro avg']['precision'])
    recall_DIET_Pretrained.append(data['macro avg']['recall'])
    f1_DIET_Pretrained.append(data['macro avg']['f1-score'])

# BERT
for i in range(1,6):
  with open('faq_chitchat_data/binary_data/binary_models_results/cv{}_BERT_result/intent_report.json'.format(i)) as f:
    data = json.load(f)
    accuracy_BERT.append(data['accuracy'])
    precision_BERT.append(data['macro avg']['precision'])
    recall_BERT.append(data['macro avg']['recall'])
    f1_BERT.append(data['macro avg']['f1-score'])

print('5-fold DIET results','\n','accuracy: ', accuracy_DIET, '\n', 'precision: ', precision_DIET, '\n', 'recall: ', recall_DIET,
      '\n', 'f1: ', f1_DIET)
print('----------------------------------------------------------------------')
print('5-fold DIET_Pretrained results','\n','accuracy: ', accuracy_DIET_Pretrained, '\n', 'precision: ', precision_DIET_Pretrained, 
      '\n', 'recall: ', recall_DIET_Pretrained, '\n', 'f1: ', f1_DIET_Pretrained)
print('----------------------------------------------------------------------')
print('5-fold BERT results','\n','accuracy: ', accuracy_BERT, '\n', 'precision: ', precision_BERT,
      '\n', 'recall: ', recall_BERT, '\n', 'f1: ', f1_BERT)
```

```python
import pandas as pd
import numpy as np
from numpy import *
comparision = pd.DataFrame(index=['F1', 'Precision', 'Recall'], columns=['CountVectorsFeat.', 'Convert Feat.', 'BERT Feat.'])
comparision.loc['F1','CountVectorsFeat.']=mean(f1_DIET)
comparision.loc['F1','Convert Feat.']=mean(f1_DIET_Pretrained)
comparision.loc['F1','BERT Feat.']=mean(f1_BERT)

comparision.loc['Precision','CountVectorsFeat.']=mean(precision_DIET)
comparision.loc['Precision','Convert Feat.']=mean(precision_DIET_Pretrained)
comparision.loc['Precision','BERT Feat.']=mean(precision_BERT)


comparision.loc['Recall','CountVectorsFeat.']=mean(recall_DIET)
comparision.loc['Recall','Convert Feat.']=mean(recall_DIET_Pretrained)
comparision.loc['Recall','BERT Feat.']=mean(recall_BERT)

comparision
```
